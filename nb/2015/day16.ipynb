```python
from aocd import get_data

puzzle_input = get_data(day=16, year=2015)
puzzle_input = puzzle_input.split("\n")
```

```python
puzzle_input[:3]
```

```python
def parse_instruction(instruction:str):
    sue, properties = instruction.split(": ", 1)

    n_sue = int(sue[4:])
    properties_dict = {}
    for prop in properties.split(", "):
        name, value = prop.split(": ")
        properties_dict[name] = int(value)

    return n_sue, properties_dict

sues = {}
for instruction in puzzle_input:
    n, props = parse_instruction(instruction)
    sues[n] = props

sues
```

Part 1

```python
scan = {
    "children": 3,
    "cats": 7,
    "samoyeds": 2,
    "pomeranians": 3,
    "akitas": 0,
    "vizslas": 0,
    "goldfish": 5,
    "trees": 3,
    "cars": 2,
    "perfumes": 1,
}

def filter_from_scan(props:dict):
    for name, value in props.items():
        if value != scan[name]:
            return False
    return True


sues_filtered = {
    n: props
    for n, props in sues.items()
    if filter_from_scan(props)
}

sues_filtered
```

Part 2

```python
scan = {
    "children": 3,
    "cats": 7,
    "samoyeds": 2,
    "pomeranians": 3,
    "akitas": 0,
    "vizslas": 0,
    "goldfish": 5,
    "trees": 3,
    "cars": 2,
    "perfumes": 1,
}

def filter_from_scan_rectified(props:dict):
    for name, value in props.items():
        if name in ("cats", "trees"):
            if  value < scan[name]:
                return False
        elif name in ("pomeranians", "goldfish"):
            if value > scan[name]:
                return False
        elif value != scan[name]:
            return False
    return True


sues_filtered = {
    n: props
    for n, props in sues.items()
    if filter_from_scan_rectified(props)
}

sues_filtered
```
