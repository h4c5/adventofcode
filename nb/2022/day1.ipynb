```python
from aocd import get_data

puzzle_input = get_data(day=1, year=2022)
```

```python
food = [[int(f) for f in food.split("\n")] for food in puzzle_input.split("\n\n")]
food
```

```python
sum_food = list(map(sum, food))
max(sum_food)
```

```python
sum(sorted(sum_food, reverse=True)[:3])
```
